# Non-Standard Orthographic Variations: Normalization Analysis

## Abstract

The following paper presents findings based on Python code driven normalization analysis of the Early Modern English Books Online project corpus of approximately 40,000 texts spanning from the late 15th Century to the turn of the 19th Century. These XML tagged documents were analyzed originally using a normalization tool called VARD, which allowed for manual, semi-automatic, and automatic normalization tagging of non-standard early modern spellings to their modern equivalents based on a set of percentage accuracy levels ranging from 35-65. With this pre-normalized data then, a script was written to extract all pertinent meta-data out of the corpora. The meta-data that was extracted included file name, date (as written in each text's header), real date (generally the same date as the previous date, however, oftentimes there were multiple forms given and the real date function narrowed down the actual date each text was written or published), author, title, a raw count of normalizations within the body of the text (taken from each XML tag as opposed to each count of the term "normalisation", and finally a raw count of the tokens/terms in each text. From there, the data was saved as a csv file within which further manual processing work took place. Finally, the dataset was analyzed again using the Python coding language, with the help of a module called Pandas which took the data frame and was manipulated into graphical visualizations to allow for a review of a broader picture of the orthographic shifts in the early modern English language. The analysis that follows suggests that there was a specific point in the orthographic history of the language at which variations were overwhelmingly abundant. 

## Introduction

Modern humanist research is rapidly expanding to include digital analysis of texts from across the centuries. This relatively new field of digital humanities allows researchers to analyze texts en masse using methods previously reserved for linguists ¬and corporations looking to further their research and development techniques based on digitally submitted product reviews. However, humanists from all research backgrounds are now using computational methods in order to understand texts on a broader scale, which have the ability to produce in-depth analytical results. Whereas prior to the utilization of such techniques, such large-scale projects would have taken significantly more time, code based analysis of texts has shrunk the amount of time to ‘read’ and analyze corpora to a mere fraction of that time. Emerging digital humanities researchers may utilize text corpora that contain significant variation in orthography, which is to be expected in historical corpora dating prior to modern orthographic standardization, and which tend to be problematic when attempting to tag parts of speech and semantic similarity, or dissimilarity for that matter, whether using automatic or manual techniques, for computational quantitative processing and subsequent intellectual qualitative analysis. 

This project holds its foundations within the Early Modern (English) Books Online, EEBO, project which has created a database of searchable texts, starting in the late 15th Century continuing officially until the turn of the 18th Century, with an additional subset of texts through the turn of the 19th Century. Out of the EEBO project stemmed the Text Creation Partnership (TCP) organization located at the University of Michigan's library since 2000. According to the TCP website:
 [t]he Text Creation Partnership creates standardized, accurate XML/SGML 
encoded electronic text editions of early print books. We transcribe and mark up 
the text from the millions of page images in ProQuest's Early English Books 
Online, Gale Cengage's Eighteenth Century Collections Online, and Readex's 
Evans Early American Imprints. This work, and the resulting text files, are jointly 
funded and owned by more than 150 libraries worldwide. All of the TCP's work 
will be released the public domain for anyone to use. (Text Creation Partnership, 2015)
As mentioned, a large aspect of the EEBO project was to turn pictures of texts on microfilm into keyword searchable digitized documents, which in theory would broadly expand potential areas of humanist research. However, ProQuest encountered a significant issue in terms of optical character recognition (OCR) due to its limitations in certain font recognition, for example early Gothic. Therefore, originally, these texts were simply input into a collection of images of the original works’ pages. According to Shawn Martin in the e-publication Digital Scholarship and Cyberinfrastructure in the Humanities: Lessons from the Text Creation Partnership, “[l]ibrarians at Michigan and Oxford believed that they could get support to capture full electronic text for at least a portion of the EEBO corpus. Thus the TCP was born.” (2007, “Background”). Given that EEBO and TCP are ground-level initiatives, there remains a need for more in depth searchability.

One issue of searchability listed above is the limitation of OCR. A further issue is that of spelling variation due to a lack of standardized spelling rules prior to the modern era. Thanks to digital humanist efforts, specifically the DREaM project and Early Modern Conversions, which seek to make the EEBO texts readily available for scholarly search queries, code-based analysis of the corpora allows for meta-data extraction from the annotated and XML tagged texts to better understand the orthographic variations, and how these variations affect searchability and organization efforts.  According to the Early Modern Conversions website’s Introduction page, “[o]ur first task, then, was to match all of the metadata files to the text files, thereby creating a new set of concatenated files that we could arrange within a searchable structure—a nice, well-organized library with shelves, catalogues, and helpful librarians to assist with queries. Our next major challenge derived from complexities introduced by spelling variance.” (Wittek, “Building the DREaM Database”, 2015). In terms of distance reading and large scale analysis, spelling variations present a significant problem due to the multitude of different spellings for any given word. Given that these spelling variations are widespread in historical texts, it seems to be nearly impossible to track the occurrence of a given term within a chronologically vast corpus. Dr. Wittek uses a famous example of spelling variation, “the name ‘Shakespeare’ has eighty different recorded spellings, including ‘Shaxpere’ and ‘Shaxberd.’ (Wittek, “Building the DREaM Database”, 2015). 

To circumvent the issues related to orthographic variation a normalization tool, VARD2,  developed by Dr. Alistair Baron at Lancaster University in the UK, searches through each text file and looks for terms that it does not recognize, i.e. are not within its real word list, and attempts to find a suitable standard modern equivalent to replace whatever variant is being questioned. At its core, VARD2 seeks to amplify the accuracy of key-word searches that are currently fruitless due to the numerous variations. VARD2 allows for manual or automatic normalizations, whether for a single text or a large corpus such as the EEBO-TCP corpus. The manual normalization process acts as a training mode, where the program can learn grammar and spelling rules based on the user’s manual input. The automatic normalization method allows for several normalization thresholds or standard percentages. The percentage given for a suggested normalization is based on the rules previously input into the program manually, or those that it automatically comes with. The thresholds used in this project are 35, 45, 50, and 65.  The words for suggested normalization are highlighted in yellow automatically by the program, which can then be moused over to see a ranked list of given suggestions and their confidence score. 
For the purposes of this project, all four of the aforementioned thresholds were analyzed using coded metadata extraction scripts, which output the requested information into neat .csv files for further computational analysis and subsequent graphical plotting. The Python coding language was used to loop through the corpus of around 40,000 texts extracting the EEBO tagged header information as well as the normalization tags within the body of each text and a total count of tokens/terms. The output columns were labeled as follows: ‘File_Name’, ‘Date’, ‘Real_Date’, ‘Title’, ‘Author’, ‘Normalization_Count’, and ‘Body_Tokens_Count’. A key distinction here is the difference between the date extracted and the real date of the text. As noted above by the TCP website, standardization and accuracy are important components of the project. However, many of the XML tagged dates within the header or root of the files are not standardized at all. There are numerous different formats ranging from Roman numerals to words such as “in the yeare of our Lorde”, from parenthetical dates “(1642)” to more complex parentheticals “(1.5.6.9.)”. Some dates are in brackets, others have multiple date tags, such as document A45145 which looked as follows: “<DATE>printed in the year, </DATE><DATE> 1671</DATE>”, which confused the script and returned an unknown date. In other instances multiple years were included due to calendric inconsistencies for example, MCDLV (1455, i.e. 1655). In situations such as these, the rarity allowed for manual manipulation of the data based on the most accurate suggestions included in the tagged headers. Other inconsistencies included fractioned years such as 1695/6, which occurred a handful of times and were manually entered into the real date field as 1696. Furthermore, due to the manual tagging process, human errors are bound to occur despite all efforts put forth in terms of verifying accuracy. One such example that occurred frequently, in the document A83860.xml for example, was [l642], which returns the output 642 because the first character in the string is alphabetical 'l' as opposed to the numerical '1' characters for which the script was searching. However, given that the corpus contains around 40,000 texts, the 2.7% calculated rate of fallibility suggests that these inconsistencies are a rather small issue that was in the end resolved by further manual processing within the extracted metadata spreadsheet.

### Viewing and Manipulating the Extracted Metadata

The process of extraction for four sets of 40,000 texts lasted approximately 12 hours for each run through which was to be expected due to the overwhelming number of normalization tags per text, as well as the tags within the header of each file. However, once the process was complete, it was a simple process to calculate the ratio of normalizations to the total token count in the body of each text file, and then from there calculate the percentage of normalizations. Furthermore, weight was calculated based on the ratio/percentage counts, which varied from 0 for any percentage below 1, 1 below 2, 2 below 3, 3 below  4, 4 below 5, and 5 for any percentage of 5 and above. Finally, a date factor was calculated based on the real date column of year values, divided into twenty year segments, except for those texts prior to 1480, which were added to the first date factor of 0. These factors increased by one per twenty year segment, except when several thousand texts were found within a decade, at which point the second half of the twenty year segment was given a .1 addition. For example, the years 1640 to 1660 were assigned the date factor 9, however due to the large number of texts within the 1640-1650 period, the texts in the second half, 1650-1660 were assigned the date factor 9.1 to differentiate early and late date factor texts. #### EEBO-TCP Metadata 35

In this section, the metadata of the VARD threshold 35 will be plotted and analyzed. As described in the previous section, all modules will be imported up front so that they will be readily available for use in the following lines of code.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import os
from pandas.tools.rplot import *
from pandas import DataFrame, read_csv
from io import StringIO
%matplotlib inline
```

The csv file is then given an object name and is then asked to be read as the given file name by the Pandas read_csv utility. The object is given the name lowCsv due to the fact that it is the lowest threshold with which this project deals. 

```python
lowCsv = pd.read_csv('EEBO_MetaData_35expanded.csv')
```

Next, a new object called sortedlowCsv is created through an automatic sorting function, in this case by the "Real_Date" column in the file. The first few lines are then shown in order to make sure that the file is indeed being read properly.

```python
sortedlowCsv = lowCsv.sort("Real_Date")
sortedlowCsv.head()
```

The above function has transformed the .csv data set into a data frame. At this point, the data frame is passed into a list comprehension function that allows several arguments to be passed simultaneously with ease. In the following chart, the ratios computed automatically by the function will appear as raw counts, the minimum ratio found within the entire data frame, the max, the mean, and the standard deviation, along with appropriate column headings. Furthermore, the real dates were automatically grouped by the each value, oftentimes returning only one text per year, but also up to several hundred per year, which makes analyzing the overall data simpler based on such groupings. 

```python
data35 = []
for name, group in sortedlowCsv.groupby("Real_Date"):
    ratios35 = group.Normalization_Count/group.Body_Tokens_Count
    data35.append([name, ratios35.count(), ratios35.min(), ratios35.max(), ratios35.mean(), ratios35.std()])
datadf35 = pd.DataFrame(data35, columns=["year","count","min","max","mean","std"])
datadf35
```

The output dataframe is somewhat truncated in order to control the size of the notebook, so we cannot see all of the information. The min, max, mean, and standard deviation would give a general idea of the scope of the given threshold level. Below, those ratios are printed.

```python
print(ratios35.min())
print(ratios35.mean())
print(ratios35.max())
print(ratios35.std())
```

The range of values in this data set go from the minimum of .005 to the max of .018, with a mean of .0115. Plotting these ratios in graphical visualizations is a good way to determine if a pattern exists without being overwhelmed by numerical values within the data frame. First we plot the min.

```python
datadf35.plot(x='year', y='min')
```

This graph shows that there is a point in time early on at which there is a dip in the ratio of normalizations to total token counts. Then around the middle of the 17th Century, minimal normalizations become common place. 

According to the Oxford English Dictionary's overview of early modern English, "Between about 1540 and 1640 there was a movement for spelling reform in England. Early advocates were Sir John Cheke and Sir Thomas Smith, who as classical philologists were conscious of the disparity in spelling between English and Latin. John Hart produced three works on the subject between 1551 and 1570 and proposed a phonetic spelling system using a number of additional symbols. In opposition to this approach, Richard Mulcaster advocated only mild reform, and there are very few improvements in his word-lists when compared with modern spelling." (Attitudes to English, 2013).

Further with Oxford, and perhaps more significantly, "In the late-fifteenth century printers began printing books written in the form of London English which had already become a kind of standard in manuscript documents. Between 1475 and about 1630 English spelling gradually became regularized." (Early modern English pronunciation and spelling, 2013). 

The timeline seems to add up with what is being shown in the above graph. However, since this shows only the minimum ratios, perhaps a plot of the maximum ratios would be more enlightening. 

```python
datadf35.plot(x='year', y='max')
```

The maximum ratios plot shows a more cohesive downward grade towards overall orthographic normalization, whether naturally or imposed by outside forces such as the publication of dictionaries or other important linguistic and philologic works. Again, since this plot only shows the maximum ratios, the next step is to plot the mean ratios in order to see the overall middle-points through the centuries. 

```python
datadf35.plot(x='year', y='mean')
```

Looking more closely at the mean ratios plot, Figure 6, suggests that the average of all ratios per grouped year mirrors the beginning portion of the minimum and maximum graphs, but does not maintain its similarity as time progresses. Indeed, the same normalization ebb occurs in the mid-17th Century through to the turn of the next, when a spike in normalizations can again be seen. 

```python
datadf35.plot(x='year', y='std')
```

Finally, the standard deviation graph shown here suggests that there are significant undulations from the mean ratios at one point approxmiately twenty-five years prior to 1500, again approximately twenty-five years after 1500, and less significantly in the beginning years of the 18th Century. However, as these analyses rely solely on the 35-threshold data, and the standard deviation graph seems to include the most pertinent and compelling information, the following sections will include graphs of the remaining three threhold levels' data to allow for comparison and in-depth analysis as to what is causing significant variations to occur at some points in history, where others are less pronounced.

### Orthographical Analysis for Presented Statistical Findings

The dates at which significant jumps in normalization occur are mainly found within the early period of the early modern English timeline, i.e. prior to 1550. In this sense based on early modern English language scholarship by historians, linguists, and philologists, it should be a fairly simple process to understand what was going on in the world at the time that would cause the corpus of over 40,000 texts to return computationally extracted metadata and statistical derivation thereof in such stark graphs here presented. A look at the one .csv file of metadata shows that the median percentage normalized for the first and second date factors, i.e. prior to 1480 through 1500, is four, which is one of the highest concentrations of such a large percentage of normalized terms. Using the EEBO Chadwyck-Healey online visual database of scans of early English books’ print forms, scanned copies of original works can be viewed and theoretically key-word searched. One such example work from Jacobus de Cessolis, actually William Caxton, dated 1474 is shown below as Figure 12. This particular text, entitled To the right noble, right excellent [and] vertuous prince George duc of Clarence Erle of warwyk and of salisburye, grete chamberlayn of Englond [and] leutenant of Irelond oldest broder of kynge Edward by the grace of god kynge of England and of frau(n)ce, your most humble seruant William Caxton amonge other of your seruantes sendes vnto you pease. helthe. loye and victorye upon your enemyes…” contains a total of 11,953 normalizations out of 259,688 tokens, giving a rounded ratio of .05 or 4.56%. Looking at the title itself,as well as the scanned image of the original
document shows from where many of the normalization issues could have stemmed.
Prior to the turn of the 16th Century, the English language was still in, what is now considered as, the Middle English period, whereby it was in large part a derivative of Indo-European Germanic. Therefore, spelling at the time was not standardized, whether in handwritten or typeset documents. According to the University of Michigan’s TCP searching guidelines for spelling, some of the main reasons that researchers may have issues searching the database of texts which are used in the presented project include abbreviations, word divisions, and macrons, which are words that have been truncated in order to fit more easily on a line of text. Macrons were “often signaled with a horizontal bar over a character in the word. For example, convenient may appear as c_venient. TCP texts transcribe such omissions using tildes, as in co~venient, rather than attempt to expand these abbreviations, which are often hard to make out. A search for convenient will not turn up this spelling.” (TCP, Early modern spelling and TCP searching). All of these situations most definitely presented problems for the EEBO-TCP digitization team in their attempts to take the texts such as Figure 12 and convert them into legible and easily searchable text files. Furthermore, these issues became clearer when entered into VARD2 for normalization. 
	
According to the online Encyclopedia Britannica’s history and characteristics of modern English, William Caxton attempted to rectify non-standard spelling with the opening of his wooden printing press in Westminster in 1476 (Orthograpy). At the beginning of the 15th Century, what is now known as the Great Vowel Shift began to spread through English language society, where many diphthongs came to be where there was previously only one vowel. Furthermore, due in part to the spread of Renaissance scholarship, Latin words were borrowed, whether through French or directly from the Latin. However, the Latin terms used were not only ‘Church Latin’, but rather also Classical Latin. (Transition from Middle English to Early Modern English). French, Latin, and Greek terms all became part of the English language. However, a closer look at the normalized XML files does not indicate that these loan words were the issue in terms of normalization ratios. One theory for the spikes or dips in normalization deviations revolves around the publication of dictionaries, with which writers could check the ‘proper’ spelling of terms. However, since The First English Dictionary by Robert Cawdrey was not published until 1604, this does not explain the significant jumps prior to that date. (Uni Due, Early Modern English). What it actually does explain, though, is the steady decline in normalizations after that date through to the 1700s. 


## Conclusions

As stated in the above analyses, many factors contributed to the variations in normalization ratios in the EEBO-TCP corpus. The advent of wooden printing that was brought to England in the late 15th Century by William Caxton significantly increased the number and quality of published documents, which at first also correlated to a rise in spelling variation, actually reigned in orthography to a certain extent, which is clearly shown in Figure 7.  Furthermore, following the publication of the first English language dictionary and subsequent similar publications, there began to appear a more cohesive orthography, which again can be clearly seen in the provided graphics. What is still unclear, however, is whether there are certain types of texts, i.e. genres, which are more likely to have significant spelling variation. Many of the texts included in the utilized corpus were church documents, political statements, personal letters, among others. As a next step, building off of the current project, an in-depth analysis of genre specific spelling variation and subsequent normalization data, extracted in a similar manner, could shine a light on what other aspects would be considered important when researching language shifts or other humanist interests. 

The digital humanities aspect of this project allowed for a broad view of an incredibly large corpus that would have been nearly impossible to study in such a manner manually. Computationally analyzed, the texts in the EEBO-TCP corpus show that there are indeed specific points in history that can be pinpointed for shifts in orthography and, perhaps also, similarly pronunciation. Since the only aspects of language that survive from the early modern English language period are these texts and not recordings of native pronunciation, it is indeed useful to analyze the corpus using such computational techniques. This project was intended to open the doors for further research on genre-specific data analysis, and has built the foundation on which this research can grow.
